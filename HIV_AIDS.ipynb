```python
# Import necessary library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# loading dataset
data = pd.read_csv('DOHMH_HIV_AIDS_Annual_Report.csv')
```

# Data Visualisation

```python
# Summary Table
summary_table = data.groupby("Borough")[
    ["HIV diagnoses", "AIDS diagnosis rate", "% viral suppression"]
].mean().reset_index()
summary_table
```

```python
# 1. Yearly Trend in HIV Diagnoses
plt.figure(figsize=(10, 6))
yearly_trend = data.groupby("Year")["HIV diagnoses"].sum()
plt.plot(yearly_trend, marker='o', color='b')
plt.title("Yearly Trend in HIV Diagnoses")
plt.xlabel("Year")
plt.ylabel("HIV Diagnoses")
plt.grid(True)
plt.show()
```

```python
# 2. HIV Diagnosis Rate by Borough
plt.figure(figsize=(12, 6))
borough_diagnosis_rate = data.groupby("Borough")["HIV diagnosis rate"].mean()

# Sort values and assign plot to ax for annotation
ax = borough_diagnosis_rate.sort_values().plot(kind='bar', color='purple')
plt.title("Average HIV Diagnosis Rate by Borough")
plt.xlabel("Borough")
plt.ylabel("HIV Diagnosis Rate")
plt.xticks(rotation=45)
# Add value labels above the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.show()
```

```python
# 3. Gender Disparity in Concurrent Diagnoses
plt.figure(figsize=(10, 6))
gender_concurrent = data.groupby("Gender")["Concurrent diagnoses"].sum()

# Plot to ax for annotation
ax = gender_concurrent.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red'])
plt.title("Concurrent Diagnoses by Gender")
plt.xlabel("Gender")
plt.ylabel("Concurrent Diagnoses")

# Add value labels above the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.show()
```

```python
# 4. Age Group Distribution in HIV Diagnoses
plt.figure(figsize=(12, 6))
age_group_diagnoses = data.groupby("Age")["HIV diagnoses"].sum().sort_values()
ax = age_group_diagnoses.plot(kind='bar', color='teal')
plt.title("HIV Diagnoses by Age Group")
plt.xlabel("Age Group")
plt.ylabel("HIV Diagnoses")
plt.xticks(rotation=45)

# Add numbers above the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

plt.show()
```

```python
# 5. Race Distribution in HIV Diagnoses
plt.figure(figsize=(8, 8))
race_diagnoses = data.groupby("Race")["HIV diagnoses"].sum()
race_diagnoses.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='viridis')
plt.title("Race Distribution in HIV Diagnoses")
plt.ylabel("")
plt.show()
```

```python
# 6. AIDS Diagnosis Rate by Borough
plt.figure(figsize=(10, 8))
borough_aids_rate = data.pivot_table(values="AIDS diagnosis rate", index="Borough", columns="Year", aggfunc="mean")
sns.heatmap(borough_aids_rate, cmap="YlGnBu", annot=True, fmt=".1f")
plt.title("AIDS Diagnosis Rate by Borough and Year")
plt.xlabel("Year")
plt.ylabel("Borough")
plt.show()
```

```python
# 7. Viral Suppression by Gender and Age
plt.figure(figsize=(50, 6))
viral_suppression = data.groupby(["Gender", "Age"])["% viral suppression"].mean().unstack()
ax = viral_suppression.plot(kind='bar', figsize=(25, 6), colormap='Set2')
plt.title("% Viral Suppression by Gender and Age")
plt.xlabel("Gender")
plt.ylabel("% Viral Suppression")
plt.xticks(rotation=45)
plt.legend(title="Age Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Add numbers above the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'center',
                   xytext = (0, 9),
                   textcoords = 'offset points')

plt.show()
```

```python
# 8. Deaths Over Time (Total and HIV-related)
plt.figure(figsize=(10, 6))
deaths_over_time = data.groupby("Year")[["Deaths", "HIV-related death rate"]].sum()
deaths_over_time.plot(marker='o', figsize=(10, 6))
plt.title("Deaths Over Time (Total and HIV-Related)")
plt.xlabel("Year")
plt.ylabel("Number of Deaths / Rate")
plt.grid(True)
plt.show()
```

```python
# 9. Correlation Matrix of Numerical Columns
plt.figure(figsize=(12, 8))
corr_matrix = data.select_dtypes(include=["float64", "int64"]).corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
# 10. Linked Care Analysis by Borough
plt.figure(figsize=(12, 6))
linked_care = data.groupby("Borough")["% linked to care within 3 months"].mean()

# Sort values and assign plot to ax for annotation
ax = linked_care.sort_values().plot(kind='bar', color='darkblue')
plt.title("Percentage Linked to Care Within 3 Months by Borough")
plt.xlabel("Borough")
plt.ylabel("% Linked to Care")
plt.xticks(rotation=45)

# Add value labels above the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.show()
```

```python
# 11. Non-HIV-Related Death Rate Across Age Groups
plt.figure(figsize=(12, 6))
age_non_hiv_death = data.groupby("Age")["Non-HIV-related death rate"].mean().sort_values()
ax = age_non_hiv_death.plot(kind='bar', color='brown')
plt.title("Non-HIV-Related Death Rate Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Non-HIV-Related Death Rate")
plt.xticks(rotation=45)

# Add numbers above the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

plt.show()
```
